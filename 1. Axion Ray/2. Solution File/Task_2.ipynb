```python
def column_wise_analysis(df):
    column_analysis = {}

    for column in df.columns:
        col_data = df[column]
        col_info = {
            'Data Type': col_data.dtype,
            'Unique Values': col_data.nunique(),
            'Missing Values': col_data.isnull().sum(),
            'Top 5 Values': col_data.value_counts().head(5).to_dict(),
            'Description': col_data.describe(include='all').to_dict()
        }
        column_analysis[column] = col_info
    return column_analysis

column_analysis = column_wise_analysis(df)
print("Column-Wise Analysis:")
for column, details in column_analysis.items():
    print(f"\n{column}:")
    for key, value in details.items():
        print(f"  {key}: {value}")
```

### Outlier Detection
- Steps:
    - Used statistical methods d IQR to detect and remove outliers.
    - Visualized outliers musing box plots.

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def detect_and_handle_outliers(df, method='capping'):
    
    numerical_cols = df.select_dtypes(include=['number']).columns  
    
    for col in numerical_cols:
        Q1 = df[col].quantile(0.25)  # First quartile (25th percentile)
        Q3 = df[col].quantile(0.75)  # Third quartile (75th percentile)
        IQR = Q3 - Q1  # Interquartile range
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        num_outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0]
        print(f"{col}: {num_outliers} outliers detected (Lower: {lower_bound:.2f}, Upper: {upper_bound:.2f})")

        if method == 'capping':
            df[col] = np.clip(df[col], lower_bound, upper_bound)  # Capping method
        elif method == 'remove':
            df = df[(df[col] >= lower_bound) & (df[col] <= upper_bound)]  # Removing outliers

    # Boxplot visualization
    plt.figure(figsize=(15, 8))
    df[numerical_cols].boxplot(rot=45)
    plt.title("Boxplot for Numerical Features (After Outlier Handling)")
    plt.show()
    
    return df

df_cleaned = detect_and_handle_outliers(df, method='remove')


```

### Generating Tags

```python
def generate_tags(row):
    tags = []
    if row['TOTALCOST'] > 10000:
        tags.append('HighRepairCost')
    if row['REPAIR_AGE'] > 500:
        tags.append("HighVehicleAge")
    if row['CAUSAL_PART_NM'] == 'ENGINE':
        tags.append('EngineFailure')
    if row['PLATFORM'] == 'A':
        tags.append('PlatformA')
    
    
    if 'BODY_STYLE' in row and pd.notna(row['BODY_STYLE']):
        tags.append(row['BODY_STYLE'])

    return ','.join(tags)
```

```python
df['Tags'] = df.apply(generate_tags, axis=1)
tag_counts = df['Tags'].str.get_dummies(sep=',').sum().sort_values(ascending=False)
print("Tag Counts:\n", tag_counts)
high_cost_repairs = df[df['Tags'].str.contains('HighRepairCost', na=False)]
print("\nBody Style vs. Failure Counts:")
body_style_failure = df.groupby('BODY_STYLE')['Tags'].apply(lambda x: x.str.split(',').explode().value_counts())
print(body_style_failure)
```
